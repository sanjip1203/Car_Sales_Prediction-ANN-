# import libaries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
```

# import dataset

```python
car_df =pd.read_csv('car_purchasing.csv',encoding='ISO-8859-1')
```

```python
car_df.head()
```

# data visualization 

```python
sns.pairplot(car_df)
```

# data preprocessing 

```python
car_df.info()
```

```python
df= car_df.drop(['customer name','customer e-mail','country'],axis=1)
df.head()#drop categorical value 
```

```python
X= df.drop(['car purchase amount'],axis=1)
X
```

```python
y= df['car purchase amount']
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
y = y.values.reshape(-1,1)
```

```python
y_scaled = scaler.fit_transform(y)
y_scaled
```

# Training model 


```python
from sklearn.model_selection import train_test_split
X_train,X_test ,y_train,y_test =train_test_split(x_scaled,y_scaled,random_state=0,test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow.keras 
from keras.models import Sequential 
from keras.layers import Dense 


model=Sequential()
model.add(Dense(5,input_dim = 5, activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
epochs_hist =model.fit(X_train,y_train ,epochs= 100,batch_size=50 ,verbose= 1, validation_split=0.2)
```

# Model Evaluation

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title("Model loss progress During Tranning")
plt.ylabel("training and validation loss ")
plt.xlabel("epoch number")
plt.legend(['training loss ','validation loss'])
```

```python
x_test = np.array([[1,60,30000,40000,800000]])
y_predict = model.predict(x_test)
```
